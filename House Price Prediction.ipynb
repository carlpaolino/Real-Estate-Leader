Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Import house price dataset

```python
# Solution: Download with SSL verification disabled (temporary fix)
import ssl
import urllib.request

# Create unverified SSL context
ssl_context = ssl.create_default_context()
ssl_context.check_hostname = False
ssl_context.verify_mode = ssl.CERT_NONE

# Monkey patch urllib to use our SSL context
original_urlopen = urllib.request.urlopen
urllib.request.urlopen = lambda *args, **kwargs: original_urlopen(*args, **kwargs, context=ssl_context)

try:
    print("üîÑ Attempting to download California housing dataset with SSL fix...")
    house_price_dataset = sklearn.datasets.fetch_california_housing()
    print("‚úÖ California housing dataset loaded successfully!")
    print(f"Dataset shape: {house_price_dataset.data.shape}")
    print(f"Target shape: {house_price_dataset.target.shape}")
    print(f"Features: {house_price_dataset.feature_names}")
except Exception as e:
    print(f"‚ùå Still having issues: {e}")
    print("Using sample data instead...")


house_price_dataset = sklearn.datasets.fetch_california_housing()

print(house_price_dataset)
```

```python
# Load the data into pandas dataset

house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns=house_price_dataset.feature_names)
```

```python
house_price_dataframe.head()
```

```python
#add the price column to dataframe

house_price_dataframe["price"] = house_price_dataset.target
house_price_dataframe.head()
#check the number of rows and columns

house_price_dataframe.shape
#get some information about the dataset
```

```python
#Check  number of rows and columns

house_price_dataframe.shape
```

```python
#Check for missing values

house_price_dataframe.isnull().sum()
```

```python
#stats measures of data

house_price_dataframe.describe()
#correlation between the variables


```

```python
#correlation of the data

correlation = house_price_dataframe.corr()
#construct a heatmap to understand the correlation

plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
#Splitting the data and target

X = house_price_dataframe.drop(['price'], axis=1)
Y = house_price_dataframe['price']
print(X)
print(Y)
#Splitting the data into training and test data

```

```python
#Splitting the data into training and test data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
#Model Training

#loading the linear regression model

```

```python
#XGB Regressor

model = XGBRegressor()
#training the model with X_train

model.fit(X_train, Y_train)
#evaluation

```

```python
# Evaluation

training_data_prediction = model.predict(X_train)
print(training_data_prediction)
```

```python
#R squared erorr

score_1 = metrics.r2_score(Y_train, training_data_prediction)


score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)
print("R squared error : ", score_1)
print("Mean Absolute Error : ", score_2)
```

```python


#plotting the actual prices and predicted prices

plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()


```
